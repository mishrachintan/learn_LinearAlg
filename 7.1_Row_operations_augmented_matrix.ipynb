```python
import numpy as np

# Equations
# 2x+2y-z = 8
# 5x+9y+z = 11
# -6x+y-2z = 4

A = np.array([
    [2,2,-1],
    [5,9,1],
    [-6,1,-2]
])

B = np.array([
    [8],
    [11],
    [4]
])
print(f"A:\n{A}")
print(f"B:\n{B}")
aug_A_B = np.column_stack((A,B))
print(f"\naug_A_B:\n{aug_A_B}")
```

```python
# Method to swap rows
print("Swapping row 1 and 2:\n")
print(f"Before swap:\n{aug_A_B}")
aug_A_B[[0,1]] = aug_A_B[[1,0]]
print(f"After swap:\n{aug_A_B}")
```

```python
print("Multiply first row by 3")
aug_A_B[0] = 3*aug_A_B[0]
print(f"After multiplication:\n{aug_A_B}")
```

```python
print("Add second row to first row and replace in second row\n")
aug_A_B[1] = aug_A_B[0]+aug_A_B[1]
print(f"After addition:\n{aug_A_B}")
```
