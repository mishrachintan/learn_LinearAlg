```python
import numpy as np

# Example 1: 2x2 matrix
A = np.array([
    [5,3],
    [4,1]
])

eigValues, eigVectors = np.linalg.eig(A)
print(f"Eigen values: {eigValues}")
print(f"\nEigen vectors (Note that these values are normalized vectors):\n{eigVectors}")
```

```python
# Example 2: 3x3 matrix
B = np.array([
    [-2,-4,2],
    [-2,1,2],
    [4,2,5]
])

eValues, eVectors = np.linalg.eig(B)
print(f"Eigen Values: {eValues}")
print(f"\nEigen Vectors (Note: these are normalized vectors):\n{eVectors}")
```
