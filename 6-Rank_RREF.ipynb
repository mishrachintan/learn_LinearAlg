```python
import numpy as np
from sympy import Matrix

matrix_pre_rref = np.array([
    [2,4,8],
    [1,2,3],
    [4,8,10]
])
rank_matrix_pre_rref = np.linalg.matrix_rank(matrix_pre_rref)
print(f"matrix_pre_rref:\n{matrix_pre_rref}")
print(f"\nRank: {rank_matrix_pre_rref}")
```

```python
# Using sympy
sympy_matrix_pre_rref = Matrix(matrix_pre_rref)
matrix_post_rref = sympy_matrix_pre_rref.rref()

print(f"Matrix pre-ref:\n{sympy_matrix_pre_rref}")
print(f"\nMatrix post-ref:\n{matrix_post_rref}")
```

```python
# Convert that sympy Matrix in numpy format
np_rref_matrix = np.array(matrix_post_rref[0], dtype=float)
print(f"Convert that sympy Matrix in numpy format:\n{np_rref_matrix}")
```

```python
print("Rank of rref:", np.linalg.matrix_rank(np_rref_matrix))
```
