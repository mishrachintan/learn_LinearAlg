```python
import numpy as np
A = np.array([[7,4], [5,7]])

print("If det(A) != 0, then inverse exists")
print("A:\n", A)
print("det(A):", np.linalg.det(A))
inv_A = np.linalg.inv(A)
print("\nHence inv(A):\n", inv_A)
```

```python
# Matrix properties
#  1. inv(A).A = A.inv(A) = I
print("Property 1: inv(A).A = A.inv(A) = I\n")
print(f"np.dot(inv_A, A):\n{np.dot(inv_A, A)} \nnp.dot(A, inv_A):\n{np.dot(A, inv_A)}")
print("\nNOTE! We see that there is a slight difference but they are essentially the same!")
```

```python
# Matrix properties
#  2. inv(AB) = inv(B).inv(A)
print("Property 2: inv(AB) = inv(B).inv(A)")
A = np.array([[2, 1], [5, 3]])
B = np.array([[1, 2], [3, 4]])
print("A:\n", A, "\nB:\n", B)
print("inv(AB):\n", np.linalg.inv(np.dot(A, B)))
print("inv(B).inv(A):\n", np.dot(np.linalg.inv(B), np.linalg.inv(A)))
```

```python
# Matrix properties
#  3. transpose(inv(A)) = inv(transpose(A))
print("Property 3: transpose(inv(A)) = inv(transpose(A))")
A_T = A.T #np.transpose(A)
inv_A = np.linalg.inv(A)
print("A:\n", A, "\nA_T:\n", A_T, "\ninv(A):\n", inv_A)
print("\ntranspose(inv(A)):\n", inv_A.T)
print("\n inv(transpose(A))):\n", np.linalg.inv(A_T))
```

```python
# Matrix properties
#  4. inv(kA) = 1/k.inv(A)
print("Property 4: inv(kA) = 1/k.inv(A)")
k=2
print("k=",k)
print("inv(kA):\n", np.linalg.inv(k*A))
print("\n(1/k)*inv(A):\n", (1/k)*np.linalg.inv(A))
```

```python
# Matrix properties
#   5. inv(inv(A)) = A
print("Property 5: inv(inv(A))=A")
print("A:\n", A)
inv_A = np.linalg.inv(A)
print("inv(A):\n", inv_A)

inv_inv_A = np.linalg.inv(inv_A)
print("inv(inv(A)):\n", inv_inv_A)
```
